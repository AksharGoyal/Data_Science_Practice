# Filtering Operations

```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt 
from ipywidgets import interact 

data = pd.read_csv('employee.csv')
data.head()
```

```python
## Simple Filtering Operations
```

```python
data[data.YearsAtCompany > 35]
```

```python
data[data['MonthlyIncome'] > 19500]
```

## Using AND and OR in filtering

```python
average = data.MonthlyIncome.mean()
data[(data.YearsInCurrentRole > 10) & (data['MonthlyIncome'] < average)]
# Use () for each condition to prevent ambiguity
```

```python
x = data[(data['Age'] > 35) | (data.MonthlyIncome > 19500)]
x[['BusinessTravel', 'Department']]
```

## Combining Filter and Groupby

```python
x = data[(data['Department'] != 'Sales') & (data['Age'] >= 40)]
x[['DailyRate', 'MonthlyIncome']].groupby(x['Department']).mean()
```

## Interactive Filerting Operations

```python
@interact
def filter_df(column = 'MonthlyIncome', salary = 19000):
    # return data[data[column] > salary].head(15).style.background_gradient('cubehelix')
    return (data[data[column] > salary][['EmployeeNumber', 'Department', 'MonthlyIncome']]).set_index('EmployeeNumber').sort_values(by = 'MonthlyIncome').head(15).style.background_gradient(cmap = 'jet')
```
