# Charts

```python
import pandas as pd 
import matplotlib.pyplot as plt
import plotly.express as px
import seaborn as sns
```

```python
data = px.data.gapminder()
data.head()
```

## Line

```python
sns.lineplot(data['year'], data['lifeExp'])
plt.show()
```

```python
px.line(data, x = 'year', y = 'lifeExp')
# We get a line for each record which makes it difficult to interpret the chart
```

```python
px.line(data, x = 'year', y = 'lifeExp', color = 'continent')
# Now this is more informative
```

```python
px.line(data, x = 'year', y = 'lifeExp', color = 'continent', line_shape = 'hv')
```

```python
px.line(data, x = 'year', y = 'lifeExp', color = 'continent', line_shape = 'vh', hover_name = 'country')
```

## Area

```python
px.area(data, x = 'year', y = 'pop', color = 'continent', line_group = 'country', hover_name = 'country')
```

## Bar

```python
px.bar(data, x = 'continent', y = 'pop')
```

```python
df = px.data.tips()
px.bar(df, x = 'sex', y = 'total_bill', color = 'smoker', barmode = 'group')
```

## Facet grids / Trellis plots / Small multiples

```python
px.bar(df, x = 'sex', y = 'total_bill', color = 'smoker', barmode = 'group', facet_row = 'time', facet_col = 'day')
```

```python
px.bar(df, x = 'sex', y = 'total_bill', color = 'smoker', barmode = 'group', facet_row = 'time', facet_col = 'day', category_orders = {'day': ['Thu', 'Fri', 'Sat', 'Sun'], 'time': ['Lunch', 'Dinner']})
```

```python
px.scatter(df, x = 'tip', y = 'total_bill', color = 'smoker', facet_row = 'time', facet_col = 'day', category_orders = {'day': ['Thu', 'Fri', 'Sat', 'Sun'], 'time': ['Lunch', 'Dinner']})
```

## violin plot

```python
px.violin(df, y = 'smoker', x = 'tip', color = 'sex', box = True, points = 'all', )
```

# Waffle Chart

A waffle chart shows progress towards a target or a completion percentage. There is a grid of small cells, of which coloured cells represent the data.

```python
from pywaffle import Waffle

data = {'phone': ['Xiomi', 'Samsung', 'Apple', 'Nokia', 'Realm'], 'stock': [44, 12, 8, 5, 3]}

df = pd.DataFrame(data)

fig = plt.figure(FigureClass = Waffle, rows = 5, values = df.stock, labels = list(df.phone))
fig.show()
```
