# Univariate, Bivariate and Multivariate Data Visualization

```python
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns

data = pd.read_csv('employee.csv')
print(data.shape)
data.tail()
```

## Univariate Analysis

```python
sns.distplot(data['Age'])
```

```python
sns.countplot(data.BusinessTravel) # Categorical Variable
```

## Bivariate Analysis

```python
sns.barplot(data['Department'], data['MonthlyIncome']) # Categorical vs Numerical
```

Employees in Sales have a better average salary

```python
sns.scatterplot(data['Age'], data['MonthlyIncome'])
```

Younger Employees tend to have less salary


## Multivariate Analysis

```python
plt.rcParams['figure.figsize'] = (18, 8)
sns.heatmap(data.corr(), annot = True, fmt = '0.2f')
plt.show()
```

If 2 columns are highly correlated, it means they are related to each other.

```python
sns.barplot(data['Department'], data['MonthlyIncome'], data['Attrition'])
```

This shows us how many employees get or don't get attrited in each department (most in HR)
